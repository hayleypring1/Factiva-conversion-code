Upload and convert txt files from Factiva to CSV, final step is to merge (2016 only)

```python
import re
import pandas as pd
from google.colab import files

# Upload your text file
uploaded = files.upload()

# Replace this with the actual filename if different
filename = list(uploaded.keys())[0]

# Read the entire text
with open(filename, "r", encoding="utf-8") as f:
    text = f.read()

# Find article start positions
article_starts = [m.start() for m in re.finditer(r"(?=^HD\s+)", text, re.MULTILINE)]
article_starts.append(len(text))  # add EOF

# Extract article blocks
articles = [text[article_starts[i]:article_starts[i+1]] for i in range(len(article_starts)-1)]

# Extract desired fields
records = []
for article in articles:
    hd = re.search(r"HD\s+(.*)", article)
    pub_date = re.search(r"PD\s+(.*)", article)
    source_name = re.search(r"SN\s+(.*)", article)
    source_code = re.search(r"SC\s+(.*)", article)
    lead_paragraph = re.search(r"LP\s+(.*?)(?=\n[A-Z]{2}\s)", article, re.DOTALL)
    full_text = re.search(r"TD\s+(.*?)(?=\n[A-Z]{2}\s|$)", article, re.DOTALL)

    records.append({
        "HD": hd.group(1).strip() if hd else None,
        "PD": pub_date.group(1).strip() if pub_date else None,
        "SN": source_name.group(1).strip() if source_name else None,
        "SC": source_code.group(1).strip() if source_code else None,
        "LP": lead_paragraph.group(1).strip().replace('\n', ' ') if lead_paragraph else None,
        "TD": full_text.group(1).strip().replace('\n', ' ') if full_text else None,
    })

# Create and save the DataFrame
df_articles = pd.DataFrame(records)
df_articles.to_csv("parsed_articlesbatch1.csv", index=False)
```
