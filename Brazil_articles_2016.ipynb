```python
import pandas as pd

# Read all uploaded files
df_2016 = pd.read_csv("Brazil_articles_2016.csv", encoding="utf-8-sig")
df_2017 = pd.read_csv("Brazil_articles_2017.csv", encoding="utf-8-sig")
df_2018 = pd.read_csv("Brazil_articles_2018.csv", encoding="utf-8-sig")
df_2019 = pd.read_csv("Brazil_articles_2019.csv", encoding="utf-8-sig")
df_2020 = pd.read_csv("Brazil_articles_2020.csv", encoding="utf-8-sig")

# Optionally tag them with the year
df_2016["year"] = 2016
df_2017["year"] = 2017
df_2018["year"] = 2018
df_2019["year"] = 2019
df_2020["year"] = 2020

# Merge all
df_all_years = pd.concat([df_2016, df_2017, df_2018, df_2019, df_2020], ignore_index=True)

# Optional: Preview
df_all_years.head()
```

```python
# Save merged data
df_all_years.to_csv("Brazil_articles_2016_2020.csv", index=False, encoding="utf-8-sig")

# Download
files.download("Brazil_articles_2016_2020.csv")
```
